```julia
using LinearAlgebra
using Plots
```

```julia
function Átomo(m,c,vx,vy,box,n)
    boxs = box
    masa = m
    carga = c
    velx = vx
    vely = vy
    ListaM = zeros(n,1)
    ListaC = zeros(n,1)
    ListaVx = zeros(n,1)
    ListaVy = zeros(n,1)
    ListaPosx = zeros(n,1)
    ListaPosy = zeros(n,1)
    for i in 1:n
        ListaM[i]=masa
        ListaC[i]=carga
        ListaVx[i]=velx
        ListaVy[i]=vely
        ListaPosx[i] = box * rand()
        ListaPosy[i] = box * rand()
    end 
    return ListaM, ListaVx, ListaVy, ListaPosx, ListaPosy 
end
```

```julia
Atom = Átomo(1,0,0,1,10,2)
```

```julia
function simulación(m,c,vx,vy,box,n)
    function Átomo(m,c,vx,vy,box,n)
    boxs = box
    masa = m
    carga = c
    velx = vx
    vely = vy
    ListaM = zeros(n,1)
    ListaC = zeros(n,1)
    ListaVx = zeros(n,1)
    ListaVy = zeros(n,1)
    ListaPosx = zeros(n,1)
    ListaPosy = zeros(n,1)
    for i in 1:n
        ListaM[i]=masa
        ListaC[i]=carga
        ListaVx[i]=velx
        ListaVy[i]=vely
        ListaPosx[i] = boxs * rand() + 5
        ListaPosy[i] = boxs * rand() + 5
    end 
    return ListaM, ListaVx, ListaVy, ListaPosx, ListaPosy 
end
    Atom = Átomo(m,c,vx,vy,box,n)
    xs = Atom[4]
    ys = Atom[5]
    scatter(xs,ys,xrange=(0,box+10),yrange=(0,box+10))
end
```

```julia
function distancia(n)
    for j in 1:n
        x_1 = Átomo(1,0,0,1,10,2)[4][j]
        y_1 = Átomo(1,0,0,1,10,2)[5][j]
        for i in 1:n
            x_2 = Átomo(1,0,0,1,10,2)[4][i]
            y_2 = Átomo(1,0,0,1,10,2)[5][i]
            distan = [sqrt((x_1-x_2)^2+(y_1-y_2)^2)]
            return distan
        end
    end
end
```

```julia
distancia(2)
```
